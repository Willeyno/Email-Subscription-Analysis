**Data transformation and Extraction**

```python
# Converting the BirthDate columns into proper date format
df['BirthDate'] = pd.to_datetime(df['BirthDate'])
df['BirthDate'].head()
```

```python
# Converting the EnrolledDate columns into proper date format
df['EnrolledDate'] = pd.to_datetime(df['EnrolledDate'])
df['EnrolledDate'].head()
df.info()
```

```python
# Extracting the day from the EnrolledDate 
df['Day'] = df['EnrolledDate'].dt.day
df.sample(10)
```

```python
# Extracting the month from the EnrolledDate 
df['Month'] = df['EnrolledDate'].dt.month
df.sample(10)
```

```python
# Extracting the year from the EnrolledDate 
df['Year'] = df['EnrolledDate'].dt.year
df.sample(5)
```

```python
# Calculating the age form the date of birth
df['Age'] = pd.to_datetime('today').year - df['BirthDate'].dt.year
df.sample(5)
```
